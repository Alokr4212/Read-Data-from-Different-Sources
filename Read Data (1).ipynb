Reading Data from different Sources 


```python
# JSON data is written key value pairs--
# Json is javascript object notataion it is is text format---

# Data Frame is two dimensional data structure in which it is store in tabular foemat data like rows and cloumn--
#---------------------------------------------------------------------------------------------------------------------------

# Convert Json format to Data Frame-------------------

import pandas as pd
from io import StringIO

Data = '''
[
  {"name": "Alok", "job": "Data Engineer"},
  {"name": "Ravi", "job": "Data Analyst"}
]
'''

df = pd.read_json(StringIO(Data))

df
```

```python
# Convert Data Frame into Json

df.to_json()
```

```python
# Orient is a parameter if we pass index it gives the index of records in json format-----


df.to_json(orient = 'index')
```

```python
# Records only give records in json format


df.to_json(orient = 'records')
```
